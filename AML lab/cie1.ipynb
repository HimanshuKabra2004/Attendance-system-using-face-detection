```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('IRIS.csv')

petal_width = data['sepal_length']
plt.subplot(1, 2, 1)
plt.hist(petal_width, bins=10, color='blue', alpha=0.7)
plt.title('Histogram Before Transformation')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

sqrt_petal_width = np.sqrt(petal_width)
print('square root of the transformation ',sqrt_petal_width)
plt.subplot(1, 2, 2)
plt.hist(sqrt_petal_width, bins=10, color='green', alpha=0.7)
plt.title('Histogram After Transformation')
plt.xlabel('Square Root (cm)')
plt.ylabel('Frequency')

plt.show()
```

```python
index = np.arange(len(petal_width))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(index, petal_width, color='blue', alpha=0.7)
plt.title('Scatter Plot Before Transformation')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.subplot(1, 2, 2)
plt.scatter(index, sqrt_petal_width, color='green', alpha=0.7)
plt.title('Scatter Plot After Transformation')
plt.xlabel('Index')
plt.ylabel('Square Root of Sepal Length (cm)') 
plt.show()
```
