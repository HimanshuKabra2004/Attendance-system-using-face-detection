```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hl(x):
    
    return np.heaviside(x,1)
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
def hl(x):
    y=x
    return y
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
def hl(x):

    
    return (1/(1+np.exp(-x)))
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
def hl(x):
    
    return (np.exp(-x)-np.exp(-x)/np.exp(-x)+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
def hl(x):
    
    return (np.tanh(x))
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
# ReLU

def hl(x):
    
    return np.where(x>0,x,0)
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
# SoftMax
def hl(x):
    
    return np.exp(x)/np.sum(np.exp(x),axis=0)
x=np.linspace(-10,10)
plt.plot(x,hl(x))
```

```python
net=np.arange(-10,10,1)

# for Hard limit

y=np.where(net>=1,1,0)
plt.plot(net,y)
plt.show()
```

```python
# for linear activation

y=np.where(net>=1,net,net)
plt.plot(net,y)
plt.show()
```
