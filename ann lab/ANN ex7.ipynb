```python
# for testing 
import numpy as np

w1 = w2 = w3 = w4 = 0.1
w5 = w6 = 0.2
n = 0.2
x1 = 0.45
x2 = 0.55
y = 0.6
epoch = 0
max_epochs = 100

outputs = []

while(epoch <= max_epochs):
    
    print(f"Epoch = {epoch}")
        
    net_h1 = w1 * x1 + w2 * x2
    y1 = 1 / (1 + np.exp(-net_h1))

    net_h2 = w3 * x1 + w4 * x2
    y2 = 1 / (1 + np.exp(-net_h2))

    net_o1 = w5 * y1 + w6 * y2
    y3 = 1 / (1 + np.exp(-net_o1))

    print(f"net_h1 = {net_h1}")
    print(f"y1 = {y1}")

    print(f"net_h2 = {net_h2}")
    print(f"y2 = {y2}")

    print(f"net_o1 = {net_o1}")
    print(f"y3 = {y3}")

    outputs.append((y1, y2, y3))
    
    e3 = y3 * (1 - y3) * (y - y3)
    print(f"e3 = {e3}")
        
    e2 = y2 * (1 - y2) * e3
    print(f"e2 = {e2}")

    e1 = y1 * (1 - y1) * e2
    print(f"e1 = {e1}")

    w1 = w1 + n * e1 * x1
    w2 = w2 + n * e1 * x2
    w3 = w3 + n * e2 * x1
    w4 = w4 + n * e2 * x2
    w5 = w5 + n * e3 * y1
    w6 = w6 + n * e3 * y2

    print(f"Updated w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}, w5 = {w5}, w6 = {w6}")
    
    epoch += 1
    print("\n")

print(f"Final output after {max_epochs} epochs: y3 = {outputs[-1][-1]}")
```
